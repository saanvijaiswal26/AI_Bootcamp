```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
s = pd.read_csv('1_The_Schedule_of_GE_to_Lok_Sabha_ii_0.csv')
s.columns = ['Sr_No', 'No_Of_Phases', 'State_UT']
s['No_Of_Phases'] = pd.to_numeric(s['No_Of_Phases'], errors='coerce')
s = s.dropna(subset=['No_Of_Phases', 'State_UT'])

plt.figure(figsize=(15, 7))
plt.bar(s['State_UT'], s['No_Of_Phases'], color='skyblue')
plt.xticks(rotation=90)
plt.xlabel('States and Union Territories')
plt.ylabel('Number of Phases')
plt.title('Number of Election Phases per State/UT')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```
